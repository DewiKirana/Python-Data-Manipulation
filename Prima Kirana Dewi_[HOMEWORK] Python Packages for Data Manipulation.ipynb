5. Buatlah kolom baru yang bernama "players category" yang berisi kategori players sebagai berikut :
- "Legend" : Jika overall bernilai lebih dari 90
- "High Class" : Jika overall bernilai dari 80 sampai 90 
- "Great" : Jika overall bernilai dari 70 sampai 80
- "Average" : Jika overall bernilai dari 60 sampai 70
- "Low Class" : Jika overall bernilai dari 50 sampai 60
- "Bottom Class" : Jika overall bernilai dibawah 50

```python
def ply_cat(x):
    if x['overall'] > 90:
        cat = 'Legend'
    elif x['overall'] > 80 and x['overall'] <=90:
        cat = 'High Class'
    elif x['overall'] > 70 and x['overall'] <=80:
        cat = 'Great' 
    elif x['overall'] > 60 and x['overall'] <=70:
        cat = 'Average'
    elif x['overall'] >=50 and x['overall'] <=60:
        cat = 'Low Class'
    elif x['overall'] <50 :
        cat = 'Bottom Class'
    return cat

data['players category'] = data.apply(lambda x: ply_cat(x), axis=1)
```

7. Bacalah 6 data yang ada di file archive, yaitu dari players_15.csv sampai players_20.csv menggunakan looping dan jadikan menjadi 1 dataframe

Hints : Lakukan concat

```python
import pandas as pd
path = 'D:/Data Science Bootcamp/Homework/Python_Library/'
list_file_name = ['players_15.csv', 'players_16.csv', 'players_17.csv', 'players_18.csv', 'players_19.csv', 'players_20.csv']

main_dataframe = pd.DataFrame(pd.read_csv(list_file_name[0]))

for i in range(1,len(list_file_name)):
    data = pd.read_csv(list_file_name[i])
    df = pd.DataFrame(data)
    main_dataframe = pd.concat([main_dataframe,df], axis = 1)
 
main_dataframe
```
